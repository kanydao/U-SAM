```python
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
sample_list = glob.glob('*.npz')
print(sample_list)
```

```python

for idx, filename in enumerate(sample_list):
    npz = np.load(filename)
    img = npz['image'].transpose()
    anno = npz['label'].transpose()
    plt.figure()
    plt.subplot(1, 2, 1)
    plt.title(f'CT Slice {idx}')
    plt.axis('off')
    plt.imshow(img, cmap='gray')
    plt.subplot(1, 2, 2)
    plt.title('Annotation')
    plt.imshow(anno, cmap='inferno',interpolation="nearest")
    plt.axis('off')
    plt.show()
```
